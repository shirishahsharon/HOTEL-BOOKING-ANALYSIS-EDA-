```python
# Select the specific column
column = df['lead_time']

# Calculate the z-scores for the column
z_scores = (column - column.mean()) / column.std()

# Set a threshold for outlier detection
threshold = 3

# Find rows where the z-scores exceed the threshold
outlier_rows = np.abs(z_scores) > threshold

# Remove the outliers by filtering the DataFrame
df_no_outliers = df[~outlier_rows]

# Create a box plot to visualize the outliers
plt.subplot(1, 2, 1)  # Create a subplot for the original data
plt.boxplot(column)
plt.title("Box Plot of {}".format('lead_time'))
plt.xlabel("Column")
plt.ylabel("Values")

plt.subplot(1, 2, 2)  # Create a subplot for the data without outliers
plt.boxplot(df_no_outliers['lead_time'])
plt.title("Box Plot of {} (Without Outliers)".format('lead_time'))
plt.xlabel("Column")
plt.ylabel("Values")

plt.tight_layout()  # Adjust the layout of subplots

# Show the plot
plt.show()
```
